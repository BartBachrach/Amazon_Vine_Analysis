```python
import pandas as pd
import numpy as np
from pathlib import Path 
```

```python
file = Path('vine_table.csv')
vine_df = pd.read_csv(file)
vine_df.head(20)
```

```python
most_reviews_df = vine_df.query('total_votes >= 20')
most_reviews_df.head()
```

```python
helpful_votes_df = most_reviews_df.query('helpful_votes / total_votes >= 0.5')
helpful_votes_df
```

```python
paid_vine_df = helpful_votes_df.query('vine == "Y"')
paid_vine_df
```

```python
unpaid_vine_df = helpful_votes_df.query('vine == "N"')
unpaid_vine_df
```

```python
count_paid_revs = len(paid_vine_df)
count_paid_revs
```

```python
five_star_df = paid_vine_df.query('star_rating == 5.0')
five_star_reviews_count = len(five_star_df)
five_star_reviews_count
```

```python
percent_five_star_paid = five_star_reviews_count / count_paid_revs
percent_five_star_paid
```

```python
count_unpaid_revs = len(unpaid_vine_df)
count_unpaid_revs
```

```python
unpaid_five_star_df = unpaid_vine_df.query('star_rating == 5.0')
len(unpaid_five_star_df)
```
